## Question 2.1

### a) Calculate the electrical effort $h_i$ and parasitic delay $p_i$ of each inverter in Figure 2

We can calculate the electrical effort $h_i$ for each stage using the formula:

$$
h_i = \frac{C_{out,i}}{C_{in,i}}
$$

For the upper chain, we have:

**INV1**

$C_{in,1} = 1$

$C_{out,1} = x$

so, we get:

$$
h_1 = \frac{x}{1} = x
$$


**INV2**

$C_{in,2} = x$

$C_{out,2} = f$

so, we get:

$$
h_2 = \frac{f}{x}
$$

For the lower chain, we have:

**INVa**

$C_{in,a} = 1$

$C_{out,a} = y$

so, we get:

$$
h_a = \frac{y}{1} = y
$$

**INVb**

$C_{in,b} = y$
$C_{out,b} = z$

so, we get:

$$
h_b = \frac{z}{y}
$$

**INVc**

$C_{in,c} = z$
$C_{out,c} = f$

so, we get:

$$
h_c = \frac{f}{z}
$$

#### Parasitic delay

Since they are all inverters we assume that they all have the same parasitic delay, which we can denote as $p_{inv}$. Therefore, we have:

$$p_1=p_2=p_a=p_b=p_c=p_{inv}$$

### b) Write an expression of the delay of each buffer chain in terms of $g_i$, $h_i$, $b_i$, and $p_i$

The delay of a path can be calculated using the formula:

$$D = \sum_{i} d_i$$

We use the following delay formula for each stage:

$$d_i = g_i h_i b_i + p_i$$

For the upper chain, we have:

$$d_1 = g_1 \, h_1 \, b_1 + p_1$$
$$d_2 = g_2 \, h_2 \, b_2 + p_2$$

$$D_{upper} = d_1 + d_2 = g_1 \, h_1 \, b_1 + p_1 + g_2 \, h_2 \, b_2 + p_2$$

Since the logical effort $g_i$ and branching effort $b_i$ are 1 for both stages, we can simplify this to:

$$D_{upper} = h_1 + p_1 + h_2 + p_2$$

For the lower chain, we have:

$$d_a = g_a \, h_a \, b_a + p_a$$
$$d_b = g_b \, h_b \, b_b + p_b$$
$$d_c = g_c \, h_c \, b_c + p_c$$

$$D_{lower} = d_a + d_b + d_c = g_a \, h_a \, b_a + p_a + g_b \, h_b \, b_b + p_b + g_c \, h_c \, b_c + p_c$$

Since the logical effort $g_i$ and branching effort $b_i$ are 1 for all stages, we can simplify this to:
$$D_{lower} = h_a + p_a + h_b + p_b + h_c + p_c$$

### c) Determine when the delays through the upper and lower buffer chains are equal

First we simplify the expressions for the delays of the upper and lower chains using the values we found in part a):

$$D_{upper} = h_1 + p_1 + h_2 + p_2 = x + \frac{f}{x} + 2p_{inv}$$

$$D_{lower} = h_a + p_a + h_b + p_b + h_c + p_c = y + \frac{z}{y} + \frac{f}{z} + 3p_{inv}$$

To find when the delays are equal, we set $D_{upper} = D_{lower}$:

$$x + \frac{f}{x} + 2p_{inv} = y + \frac{z}{y} + \frac{f}{z} + 3p_{inv}$$

Simplifying, we get:

$$x + \frac{f}{x} = y + \frac{z}{y} + \frac{f}{z} + p_{inv}$$

### a) Write down expressions for the path efforts, $F = \prod{(g_i h_i)}$, of each buffer chain

The path effort $F$ is calculated as the product of the logical efforts $g_i$ and electrical efforts $h_i$ for each stage in the path.

For the upper chain, we have:
$$F_{upper} = g_1 h_1 \cdot g_2 h_2 = 1 \cdot x \cdot 1 \cdot \frac{f}{x} = f$$

For the lower chain, we have:
$$F_{lower} = g_a h_a \cdot g_b h_b \cdot g_c h_c = 1 \cdot y \cdot 1 \cdot \frac{z}{y} \cdot 1 \cdot \frac{f}{z} = f$$

### b) Choose $h_a$ and $h_c$ to give least delay of the lower buffer chain, and calculate $h_1$

The best stage effort for a chain is given by the formula:
$$\hat{f} = F^{1/N}$$
where $N$ is the number of stages in the chain.

For the lower chain, minimum delay is obtained when all stage efforts are equal.
The best stage effort for the lower chain (3 stages) is given by:
$$\hat{f} = F_{lower}^{1/3} = f^{1/3}$$

So we set $h_a = h_b = h_c = \hat{f}$. This gives us:
$$h_a = h_b = h_c = f^{1/3}$$


Since INV2 should be equal in delay to INVb, we have:
